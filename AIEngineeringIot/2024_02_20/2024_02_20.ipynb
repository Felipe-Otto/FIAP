Arrumando os quadrantes da imagem

```python
image_goku = cv2.imread('/content/images/gokuinvertido.jpg')
image_goku = cv2.cvtColor(image_goku, cv2.COLOR_BGR2RGB)
height = image_goku.shape[0]
width = image_goku.shape[1]
plt.imshow(image_goku)
plt.show()
Q1 = image_goku[int(height/2):height, int(width/2):width]
Q2 = image_goku[int(height/2):height, 0:int(width/2)]
Q3 = image_goku[0:int(height/2), int(width/2):width]
Q4 = image_goku[0:int(height/2), 0:int(width/2)]
image_goku_right = image_goku.copy()
image_goku_right[0:int(height/2), int(width/2):width] = Q2
image_goku_right[0:int(height/2), 0:int(width/2)] = Q1
image_goku_right[int(height/2):height, int(width/2):width] = Q4
image_goku_right[int(height/2):height, 0:int(width/2)] = Q3
plt.imshow(image_goku_right)
plt.show()
```

```python
image_ball = cv2.imread('/content/images/Capture.png')
image_ball = cv2.cvtColor(image_ball, cv2.COLOR_BGR2RGB)
plt.imshow(image_ball)
plt.show()
for column in range(0, image_ball.shape[0]):
  for row in range(0, image_ball.shape[1]):
    if image_ball[column, row, 0] < 250:
      image_ball[column, row] = (0, 0, 0)
    else:
      image_ball[column, row] = (255, 255, 255)
plt.imshow(image_ball)
plt.show()
```
